# Inner join

An inner join returns only the rows where the key exists in both DataFrames. Rows that don't find a match in the other table are dropped entirely.

Inner join = intersection. Only rows present in both tables survive. It's the strictest join — safest when you only want complete, matched data, but you can silently lose rows if the keys don't align perfectly.

```python
df_employees = pd.read_csv('data/dummy_employees.csv')
df_reviews = pd.read_csv('data/performance_review.csv')

print("Employees total:", len(df_employees)) 
print("Reviews total:", len(df_reviews))

# inner join on Name (default how="inner")
df_inner = df_employees.merge(df_reviews, on='Name', how='inner')
display(df_inner)

# if the reviews file used "EmployeeName" instead of "Name"
# df_employees.merge(df_reviews, left_on="Name", right_on="EmployeeName", how="inner")
```

# Left join

A left join keeps all rows from the left DataFrame, and brings in matching data from the right. Where there's no match, the right-side columns are filled with NaN.

Left join = "Give me everything from the left table, and attach whatever you can find from the right." The left table is always fully preserved — it's the safest default when you're enriching a primary dataset with supplementary data.

```python
df_left = df_employees.merge(df_reviews, on='Name', how='left')
display(df_left)
```

```python
# Practical: flag who is eligible for promotion (False for unreviewed)
df_left["PromotionEligible"] = df_left["PromotionEligible"].fillna(False)
display(df_left[["Name", "Salary", "PerformanceRating", "PromotionEligible"]])
```

Right join — rarely used in practice. Anything a right join does can be rewritten as a left join by swapping the order of the DataFrames.

Outer join — used, but less commonly than inner/left. It's the most permissive join — keeps all rows from both sides, with NaN wherever there's no match.
